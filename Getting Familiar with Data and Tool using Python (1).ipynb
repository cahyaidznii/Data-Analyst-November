selanjutnya kita melakukan analisis pengecekan untuk variabel "NTB_Tag" 

```python
#turn date object to datetime
df['GCIF_OPNDT'] = pd.to_datetime(df['GCIF_OPNDT'])
```

```python
from datetime import datetime

df['DOpen']  = (datetime(2021, 9, 30) - df['GCIF_OPNDT']).dt.days.astype(int)
```

```python
print('NTB')
print(df[df['NTB_Tag'] == 'NTB']['DOpen'].describe().astype(int))

print('Existing')
print(df[df['NTB_Tag'] == 'Existing']['DOpen'].describe().astype(int))
```

dapat terlihat bahwa terdapat ketidak cocokan tagging NTB dan Existing karena ada DOpen yang dibawah 365 tetapi masuk dalam kategori Existing. Sehingga kita harus melakukan perbaikan Tagging nya

```python
temp1 = df[df['NTB_Tag'] == 'Existing']
temp2 = temp1[temp1['DOpen'] < 365]
```

```python
df.loc[df['GCIF_NO'].isin(temp2['GCIF_NO']) ==True, 'NTB_Tag'] = 'NTB'
df.head()
```

```python
print('NTB')
print(df[df['NTB_Tag'] == 'NTB']['DOpen'].describe().astype(int))

print('Existing')
print(df[df['NTB_Tag'] == 'Existing']['DOpen'].describe().astype(int))
```
